```python
import numpy as np
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
image_url = "https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg"
```

```python
# Read the image from url
img = Image.open(BytesIO(requests.get(image_url).content))
```

```python
# Save image using python
img.save('msTaken.jpg')
plt.imshow(Image.open('msTaken.jpg')); # Display saved file
```

```python
# Display file loaded from url
import matplotlib.pyplot as plt
# img.show() #will open image in imagemagik window
plt.imshow(img);
```

```python
# Shape of image
print("Shape of Image: height-%d, width-%d" %img.size)
```

```python
# Resize
imgr = img.resize((224, 224))
# imgr.show()
plt.imshow(imgr);
```

```python
# Save resized image
imgr.save('msTokenResized.jpg')
```
