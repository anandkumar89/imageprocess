```python
import numpy as np
from skimage.transform import rescale, resize
from skimage.io import imread, imsave

image_url = "https://assets.myntassets.com/h_640,q_90,w_480/v1/assets/images/2354027/2018/1/16/11516097831111-MsTaken-Women-Navy-Blue-Striped-Top-4731516097830960-1.jpg"
```

```python
# Download image using python

def download(fname, image_url):
        imsave(fname, imread(image_url))
        print("successfully downloaded : " + fname + "!")
        
download("ms_taken.jpg", image_url)
```

```python
# Read the image
def url_to_image(url):
    return imread(url)

# reading url
img = url_to_image(image_url)
```

```python
# Display
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(img, );
```

```python
# Shape of image
print("Shape of Image: height-%d, width-%d, color channel-%d" %img.shape)
```

```python
# Resize
imgr = resize(img, (224, 224), anti_aliasing=True, mode='constant')
plt.imshow(imgr);
```

```python
# Save resized image
imsave("MsToken_Resized.jpg", imgr)
```
